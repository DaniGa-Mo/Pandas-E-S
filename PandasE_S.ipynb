#**Aula 1 - Leyendo archivos CSV**

##Importando archivo CSV

```python
# Impoetación de la libreria pandas
import pandas as pd
```

```python
# Lectura del archivo csv (valores separados por coma) asignado a una variable
# datos.
datos = pd.read_csv('superstore_data.csv')
```

```python
# Inspección de las primeras 5 filas del dataframea.
datos.head()
```

```python
# Podemos usar una biblioteca llamada chardet para detectar el encoding de un archivo CSV.
import chardet

# Luego escribimos el siguiente bloque de código.
with open('superstore_data.csv', 'rb') as f:
  result = chardet.detect(f.read())
  print(result)
```

```python
# Inpeccionamos el segundo archivo csv, pero esta vez ceparado por punto y coma.
datos_punto_coma = pd.read_csv('superstore_data_punto_coma.csv')
```

```python
# Inspeccionamos las primeras 5 filas del dataframe encontrando con una mala visualizacion.
datos_punto_coma.head()
```

##Parámetros de la función CSV

Documentación: [pandas.read_csv](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html)

```python
# Con el parametro sep y especificando que el separador es ';', corregimos el error.
datos_punto_coma = pd.read_csv('superstore_data_punto_coma.csv', sep = ';')

# Inspeccionamos las primeras 5 filas.
datos_punto_coma.head()
```

```python
# Existe una funcion dentro de los parametros llamada nrows que nos permite
# ingrsar la cantidad de parametros que deseamos imprimir.

datos_primeras_filas = pd.read_csv('superstore_data.csv', nrows = 5)
datos_primeras_filas
```

```python
# Podemos restringir la cantidad de columnas importadas del datafram con la funcion
# usecols dentro de los parametros de .read_csv().
datos_seleccionados = pd.read_csv('superstore_data.csv', usecols = ['Id','Year_Birth','Income'])
datos_seleccionados
```

```python
# También se puede seleccionar con la posición de la columna.
datos_seleccionados = pd.read_csv('superstore_data.csv', usecols = [0,1,4])
datos_seleccionados
```

##Escribiendo archivo CSV


```python
# Para escribir un archivo csv usamos la siguiente función.
datos_seleccionados.to_csv('cliente_mercado.csv')
```

```python
# Realizamos lectura de nuestro archivo creado.
datos_cliente = pd.read_csv('cliente_mercado.csv')
# Al imprimir el archivo nos damos cuenta de que autmaticamente a creado una
# columna adicional llamada Unnamed que es una copia del indice.
datos_cliente
```

```python
# Corregiremos la creación de la columna con la siguienfe funcion de python.
# Volvemos a crear nuestro archivo csv.
datos_seleccionados.to_csv('datos_mercado.csv', index=False)
```

```python
# Volvemos a hacer lectura de nuestro archivo con ya corregido.
datos_mercado = pd.read_csv('datos_mercado.csv')
datos_mercado
```

#**Aula 2 - Leyendo archivos Excel**

##Importando archivo Excel

```python
import pandas as pd
```

```python
archivo = 'emisiones_CO2.xlsx'
```

```python
datos_co2 = pd.read_excel(archivo)
```

```python
datos_co2.head()
```

```python
# Esta funcion me permite ver la cantidad y los nombre se las hojas que posee el archivo Excel.
pd.ExcelFile(archivo).sheet_names
```

##Parámetros de la función read_excel

Documentación: [Libreria_python](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_excel.html)

```python
# Para seleccionar otra hoja del archivo excel
percapita = pd.read_excel(archivo, sheet_name='emisiones_percapita')
```

```python
# .head() me permite ver las primeras 5 filas de la hoja.
percapita.head()
```

```python
# .tail() me permite ver los ultimas 5 filas de la hoja.
percapita.tail()
```

```python
# .sample() me permite optener una cantidad de filas que yo desee de forma aleatora ingresando un numero.
percapita.sample(5)
```

```python
# Analizamos la última hoja del archivo excel.
fuentes = pd.read_excel(archivo, sheet_name='fuentes')
```

```python
fuentes.tail()
```

```python
# Para seleccionar una cantidad de columnas.
intervalo = pd.read_excel(archivo, sheet_name='emisiones_C02', usecols='A:D')
```

```python
intervalo
```

```python
# Agregamos además un filtro de filas.
intervalo_2 = pd.read_excel(archivo, sheet_name='emisiones_C02', usecols='A:D', nrows=10)
```

```python
intervalo_2
```

## Escribiendo archivo Excel

```python
# Importación de archivo excel.
intervalo.to_excel('co2_percapita.xlsx', index=False)
```

```python
# Visualización.
pd.read_excel('co2_percapita.xlsx')
```
